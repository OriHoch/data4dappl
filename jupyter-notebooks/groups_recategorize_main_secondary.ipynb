````python
import getpass, os

CKAN_API_KEY = os.environ.get('CKAN_API_KEY') or getpass.getpass("CKAN_API_KEY")
````

````python
CKAN_URL='https://www.odata.org.il'
````

````python
from dataflows import Flow, load, printer

all_groups_recategorized_main_and_secondary = Flow(
    load('data/all_groups 2019-12-21_recategorized_main_and_secondary.csv'),
    printer(tablefmt='html'),
).results()
````

````python
import requests

last_group_name = None

for recategorize_group in all_groups_recategorized_main_and_secondary[0][0]:
    group = requests.get(CKAN_URL+'/api/3/action/group_show?id={}&include_datasets=true'.format(recategorize_group['id']), headers={'Authorization': CKAN_API_KEY}).json()['result']
    set_main = False
    set_secondary = False
    for extra in group['extras']:
        if extra['key'] == 'entity_main_type':
            extra['value'] = recategorize_group['entity_main_type']
            set_main = True
        elif extra['key'] == 'entity_secondary_type':
            extra['value'] = recategorize_group['entity_secondary_type']
            set_secondary = True
    if not set_main:
        group['extras'].append({'key': 'entity_main_type', 'value': recategorize_group['entity_main_type']})
    if not set_secondary:
        group['extras'].append({'key': 'entity_secondary_type', 'value': recategorize_group['entity_secondary_type']})
    result = requests.post(CKAN_URL+'/api/3/action/group_update', json=group, headers={'Authorization': CKAN_API_KEY}).json()
    assert result['success']
    print(group['name'])
#     if last_group_name == 'foi-201908-139':
#         break
    last_group_name = group['name']
````
