````python
import getpass, os

CKAN_API_KEY = os.environ.get('CKAN_API_KEY') or getpass.getpass("CKAN_API_KEY")
````

````python
CKAN_URL='https://www.odata.org.il'
````

````python
from dataflows import Flow, load, printer

all_groups_to_add = Flow(
    load('data/all_groups 2019-12-21_to_add.csv'),
    printer(tablefmt='html'),
).results()
````

````python
import requests
import traceback
from slugify import slugify

skip_to = 'hpnyqs-khbrh-lbytvkh-b-m'
ok_to_iterate = False
for new_group in all_groups_to_add[0][0]:
    if not new_group['name']:
        new_group['name'] = slugify(new_group['display_name'])
    new_group['name'] = new_group['name'].lower()
    if skip_to == str(new_group['name']):
        ok_to_iterate = True
        continue
    if not ok_to_iterate: continue
    print(new_group['display_name'])
    print(new_group['name'])
    res = requests.post(
        'https://www.odata.org.il/api/3/action/group_create', 
        json={
            'name': new_group['name'],
            'title': new_group['display_name'],
            'extras': [
                {'key': 'entity_main_type', 'value': new_group['entity_main_type']},
                {'key': 'entity_secondary_type', 'value': new_group['entity_secondary_type']}
            ]
        },
        headers={'Authorization': CKAN_API_KEY}
    ).json()
    try:
        assert res['success'], str(res)
    except Exception:
        traceback.print_exc()
````
